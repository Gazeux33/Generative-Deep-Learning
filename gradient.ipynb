```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 4*x -5
```

```python
line_x = np.linspace(-100, 100, 1000) 

plt.plot(line_x, f(line_x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
print(f"f(3)={f(3)}")
print(f"f(2)={f(2)}")
```

```python
x = tf.Variable(2.0)
print("f(x)=",f(x))
```

```python
with tf.GradientTape() as tape:
    x = tf.Variable(3.0)
    y = f(x)
dy = tape.gradient(y,x)
print(dy)
```
